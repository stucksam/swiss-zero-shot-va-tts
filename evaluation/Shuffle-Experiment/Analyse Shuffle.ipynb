```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
df = pd.read_excel("shuffle_evaluation.xlsx", sheet_name="Overview")
```

```python
df.head()
```

```python
metrics = ['wer_lower', 'cer_lower', 'bert_score', 'bleu_score_lower']

for metric in metrics:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, x='Type', y=metric)
    plt.title(f'Distribution of {metric} by Type')
    plt.ylabel(metric)
    plt.xlabel('Type')
    plt.show()
```

```python
for metric in metrics:
    plt.figure(figsize=(8, 5))
    sns.violinplot(data=df, x='Type', y=metric, inner='quartile')
    plt.title(f'Distribution of {metric} by Type (Violin Plot)')
    plt.ylabel(metric)
    plt.xlabel('Type')
    plt.show()
```

```python
sns.pairplot(df, hue='Type', vars=metrics, height=2.5)
plt.show()
```

```python

for metric in metrics:
    plt.figure(figsize=(8, 5))
    sns.barplot(data=df, x='Type', y=metric, ci="sd", estimator=np.mean)
    plt.title(f'Mean {metric} by Type with Error Bars')
    plt.ylabel(metric)
    plt.xlabel('Type')
    plt.show()
```

```python
# Create a 2x2 grid for the plots
fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True)

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Plot each metric as a violin plot
for i, metric in enumerate(metrics):
    sns.violinplot(data=df, x='Type', y=metric, inner='quartile', ax=axes[i])
    axes[i].set_title(f'Distribution of {metric}')
    axes[i].set_xlabel('Type')
    axes[i].set_ylabel(metric)

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```
