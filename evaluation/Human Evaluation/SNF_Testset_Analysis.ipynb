```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
df_eval = pd.read_excel("SNF_Test_Sentences.xlsx", sheet_name="SNF Test Samples")
df_quality = pd.read_csv("speaker_quality_distro.csv", sep=",")
```

```python
# Display the distribution of dialect_region
plt.figure(figsize=(10, 6))
sns.countplot(data=df_eval, x='dialect_region', palette='viridis')
plt.title('Distribution of Dialect Regions')
plt.xlabel('Dialect Region')
plt.ylabel('Number of Speakers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by dialect_region and gender, get counts
grouped_df = df_eval.groupby(['dialect_region', 'gender']).size().reset_index(name='count')

# Plotting the distribution of dialect_region and gender
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_df, x='dialect_region', y='count', hue='gender', palette='coolwarm_r')
plt.title('Distribution of Dialects and Gender')
plt.xlabel('Dialect Region')
plt.ylabel('Number of Speakers')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram of the overall gender distribution

short_speaker_ids = {sid: f"S{idx+1}" for idx, sid in enumerate(df_eval['client_id'].unique())}
df_eval['short_speaker_id'] = df_eval['client_id'].map(short_speaker_ids)

plt.figure(figsize=(8, 4))
sns.histplot(data=df_eval, x='gender', stat='count', discrete=True)
plt.title('Overall Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Histogram of gender distribution grouped by speaker_id
plt.figure(figsize=(8, 4))
sns.histplot(data=df_eval, x='gender', hue='short_speaker_id', stat='count', discrete=True, multiple='stack')
plt.title('Gender Distribution by Speaker ID (Separated)')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Speaker ID')
plt.show()

# Histogram of speaker_id distribution (total)
plt.figure(figsize=(8, 4))
sns.histplot(data=df_eval, x='short_speaker_id', stat='count', discrete=True)
plt.title('Overall Speaker ID Distribution')
plt.xlabel('Speaker ID')
plt.ylabel('Count')
plt.show()

# Histogram of speaker_id distribution grouped by gender
plt.figure(figsize=(8, 4))
plot = sns.histplot(data=df_eval, x='short_speaker_id', hue='gender', stat='count', discrete=True, multiple='stack')
plt.title('Speaker ID Distribution by Gender (Separated)')
plt.xlabel('Speaker ID')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# 1. Bar plot of gender counts grouped by client_id
plt.figure(figsize=(8, 4))
sns.countplot(data=df_eval, x='short_speaker_id', hue='gender')
plt.title('Gender Distribution by Client ID')
plt.xlabel('Client ID')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()


# Grouping data by gender and age
grouped_df = df_eval.groupby(['age', 'gender']).size().reset_index(name='count')

# Bar plot of male and female speakers in relation to their age
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_df, x='age', y='count', hue='gender', palette='coolwarm_r')
plt.title('Distribution of Male and Female Speakers by Age')
plt.xlabel('Age')
plt.ylabel('Number of Speakers')
plt.legend(title='Gender')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()


# 3. Histograms of age distribution grouped by client_id
plt.figure(figsize=(8, 4))
sns.histplot(data=df_eval, x='age', hue='short_speaker_id', multiple='stack', kde=True)
plt.title('Age Distribution Grouped by Client ID')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Client ID')
plt.xticks(rotation=45)
plt.show()

# 4. Swarm plot for age distribution by client_id and gender (now treats age as categorical)
plt.figure(figsize=(8, 4))
sns.swarmplot(data=df_eval, x='short_speaker_id', y='age', hue='gender', dodge=True)
plt.title('Age and Gender Distribution by Client ID')
plt.xlabel('Client ID')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

```python
# Determine target counts per gender
# Custom order for age groups
age_order = ['teens', 'twenties', 'thirties', 'forties', 'fifties', 'sixties']
df_eval['gender'] = df_eval['gender'].astype('category')
df_eval['dialect_region'] = df_eval['dialect_region'].astype('category')

# Perform stratified sampling: 9 samples per dialect_region
# balanced_sample = df_eval.groupby('dialect_region', group_keys=False).apply(lambda x: x.sample(9, random_state=42))
balanced_sample = df_eval.groupby('dialect_region', group_keys=False).apply(lambda x: x.sample(6, random_state=42))


# Now visualize with a bar plot of male and female speakers in relation to their age
plt.figure(figsize=(10, 6))
sns.countplot(data=balanced_sample, x='dialect_region', hue='gender', palette='coolwarm_r')
plt.title('Balanced Distribution of Male and Female Speakers by Custom Dialect Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Speakers')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.show()

# Display the distribution of dialect_region
plt.figure(figsize=(10, 6))
sns.countplot(data=balanced_sample, x='dialect_region', palette='viridis')
plt.title('Distribution of Dialect Regions')
plt.xlabel('Dialect Region')
plt.ylabel('Number of Speakers')
plt.xticks(rotation=45)
plt.show()
```

```python
balanced_sample = balanced_sample.sample(frac=1, random_state=42).reset_index(drop=True)
print(len(balanced_sample))
balanced_sample.head()
```
