```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("SNF_Test_Sentences.xlsx")
```

```python
df.head()
```

```python
grouped_df = (
    df.groupby(['client_id', 'dialect_region', 'age', 'gender'])
    .agg(num_sentences=('sentence_id', 'count'))  # Count number of sentences
    .reset_index()  # Reset index to turn it back into a standard DataFrame
)
grouped_df
```

```python
# Distribution of sentences spoken by speakers
plt.figure(figsize=(8, 5))
bins = range(1, 9)  # Create bins 1 to 7 with an extra to account for right edge
plt.hist(grouped_df['num_sentences'], bins=bins, edgecolor='black', align='left', rwidth=0.8)

# Format the x-axis ticks
plt.xticks(range(1, 8))

# Add labels and title
plt.xlabel('Number of Sentences Spoken', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Number of Sentences Spoken by Speakers', fontsize=14)
plt.yticks(range(0, 19, 2))
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Count the number of samples per dialect region
dialect_counts = df['dialect_region'].value_counts()

# Plot the dialect region distribution
plt.figure(figsize=(12, 6))
dialect_counts.plot(kind='bar', color='seagreen')
plt.title('Distribution of Dialect Regions', fontsize=16)
plt.xlabel('Dialect Region', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Count the number of male and female speakers
gender_counts = df['gender'].value_counts()

# Plot the gender distribution
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Distribution of Female to Male Speakers', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
grouped_data = df.groupby(['dialect_region', 'gender']).size().unstack(fill_value=0)

grouped_data.plot(kind='bar', figsize=(10, 6), edgecolor='black', width=0.8)

# Customize the plot
plt.title('Gender Distribution Across Dialect Regions', fontsize=14)
plt.xlabel('Dialect Region', fontsize=12)
plt.ylabel('Number of Samples', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Gender', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)


# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Count unique speakers in each dialect region
unique_speakers = df.groupby('dialect_region')['client_id'].nunique()

# Plot
plt.figure(figsize=(10, 6))
unique_speakers.plot(kind='bar', edgecolor='black', width=0.8)

# Customize the plot
plt.title('Number of Unique Speakers per Dialect Region', fontsize=14)
plt.xlabel('Dialect Region', fontsize=12)
plt.ylabel('Number of Unique Speakers', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Count unique speakers in each dialect region# Explicitly order the age groups
age_order = ['teens', 'twenties', 'thirties', 'fourties', 'fifties', 'sixties']

num_age_groups = df.groupby('age')['client_id'].nunique()
num_age_groups = num_age_groups.reindex(age_order)

# Plot
plt.figure(figsize=(10, 6))
num_age_groups.plot(kind='bar', edgecolor='black', width=0.8)

# Customize the plot
plt.title('Number of Unique Speakers per Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Number of Unique Speakers', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
